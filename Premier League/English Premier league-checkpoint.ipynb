```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import data

```python
shot_creation = pd.read_csv("england_premier_league_squad_goal_shot_creation_22.csv", sep =';')
defensive_actions = pd.read_csv("england_premier_league_squad_defensive_actions_22.csv", sep =';')
goalkeeping = pd.read_csv("england_premier_league_squad_goalkeeping_22.csv", sep =';')
passing_stat = pd.read_csv("england_premier_league_squad_passing_stats_22.csv", sep =';')
possession_stat = pd.read_csv("england_premier_league_squad_possession_22.csv", sep =';')
shooting_stat = pd.read_csv("england_premier_league_squad_shooting_22.csv", sep =';')
league_stat = pd.read_csv("england_premier_league_stats_22.csv", sep =';')
league_table = pd.read_csv("england_premier_league_table_22.csv", sep =';')
```

## Visualizing Data and building some intuitions

```python
league_table.sort_values(["squa"])
```

```python
plt.figure(figsize=[15,6])
plt.scatter(goalkeeping["SaveRt"], league_table["GA"], color='blue')

plt.title('Correlation between Goalkeeping Save Rate and Goals conceded', fontweight="bold")
plt.xlabel('Save Rate', fontweight="bold")
plt.ylabel('Goals Conceded', fontweight="bold");
```

```python
plt.figure(figsize=[15,6])
plt.scatter(passing_stat["CmpShRt"], league_stat["GlsTot"], color='blue')
plt.scatter(passing_stat["CmpMedRt"], league_stat["GlsTot"], color='green')
plt.scatter(passing_stat["CmpLngRt"], league_stat["GlsTot"], color='red')

plt.title('Correlation between Completed passes percentage and Team Goals', fontweight="bold")
plt.xlabel('Completed Passes%', fontweight="bold")
plt.ylabel('Goals Scored', fontweight="bold");
plt.legend(['Short', 'Medium', 'Long'], loc="lower right");
```

* It's obvious that Long completed passes are very efficient in getting more team goals

## A prevalent debate have always been how much does the location where you apply pressure on your opposition matter?
   * We decide to answer this question based on the data we have in hand

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, figsize=[16,9])
ax1.scatter(defensive_actions["Def3rdPres"], league_stat["Poss"], color='blue')
ax1.scatter(defensive_actions["Mid3rdPres"], league_stat["Poss"], color='green')
ax1.scatter(defensive_actions["Att3rdPres"], league_stat["Poss"], color='red')
ax1.set_title('Relevance of Pressure application position', fontweight="bold")
ax1.set_ylabel('Team Possession', fontweight="bold")
ax1.legend(['Defensive 3rd', 'Middle 3rd', 'Attacking 3rd'], loc="lower right");



ax2.scatter(defensive_actions["Def3rdPres"], league_stat["GlsTot"], color='blue')
ax2.scatter(defensive_actions["Mid3rdPres"], league_stat["GlsTot"], color='green')
ax2.scatter(defensive_actions["Att3rdPres"], league_stat["GlsTot"], color='red')
ax2.set_ylabel('Goals scored', fontweight="bold")
ax2.legend(['Defensive 3rd', 'Middle 3rd', 'Attacking 3rd'], loc="lower right");


ax3.scatter(defensive_actions["Def3rdPres"], league_stat["CrdYel"], color='blue')
ax3.scatter(defensive_actions["Mid3rdPres"], league_stat["CrdYel"], color='green')
ax3.scatter(defensive_actions["Att3rdPres"], league_stat["CrdYel"], color='red')
ax3.set_ylabel('No. of yellow cards', fontweight="bold")
ax3.legend(['Defensive 3rd', 'Middle 3rd', 'Attacking 3rd'], loc="lower right");


ax4.scatter(defensive_actions["Def3rdPres"], league_stat["CrdRed"], color='blue')
ax4.scatter(defensive_actions["Mid3rdPres"], league_stat["CrdRed"], color='green')
ax4.scatter(defensive_actions["Att3rdPres"], league_stat["CrdRed"], color='red')
ax4.set_xlabel('Pressure', fontweight="bold")
ax4.set_ylabel('No. of Red Cards', fontweight="bold")
ax4.legend(['Defensive 3rd', 'Middle 3rd', 'Attacking 3rd'], loc="lower right");
```

* From this analysis, we can see a clear correllation between the area of pitch where pressure was applied with other important variable in football.

```python
sorted_by_goals = league_stat.sort_values(["GlsTot"], ascending=False).head()
conceded_goal_sorted = league_table.sort_values(["GA"], ascending=False).head()
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].bar(sorted_by_goals["Squad"], sorted_by_goals["GlsTot"], color='green')
ax[0].set_title('Top 5 goal scoring Team', fontweight="bold")
ax[0].set_ylabel('No. of Goals', fontweight="bold")
ax[0].tick_params(axis='x', rotation=90)
mean_goals_scored = np.mean(league_stat["GlsTot"])
ax[0].axhline(mean_goals_scored, color="black", linestyle="--")

ax[1].bar(conceded_goal_sorted["Squad"], conceded_goal_sorted["GA"], color='red')
ax[1].set_title('Top 5 goal conceding Team', fontweight="bold")
ax[1].tick_params(axis='x', rotation=90)
mean_goals_conceded = np.mean(league_table["GA"])
ax[1].axhline(mean_goals_conceded, color="black", linestyle="--");
```

* This shows the top 5 teams that scored more and conceded more in the season. The dotted lines shows the total average for the data under inspection.
